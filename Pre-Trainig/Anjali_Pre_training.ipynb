```python
from datasets import load_dataset
import os
```

```python
data=load_dataset('text',data_files='/content/Pretraining.txt')
```

```python
from transformers import (
    GPT2TokenizerFast,
    GPT2Config,
    GPT2LMHeadModel,
    DataCollatorForLanguageModeling,
    Trainer,
    TrainingArguments
)
```

```python
tokenizer=GPT2TokenizerFast.from_pretrained("gpt2")
```

```python
tokenizer.pad_token=tokenizer.eos_token
def tokenize_function(examples):
    return tokenizer(examples["text"])
tokenized_data=data.map(tokenize_function,batched=True,num_proc=4,remove_columns=["text"])
```

```python
config=GPT2Config(
    vocab_size=tokenizer.vocab_size,
    bos_token_id=tokenizer.bos_token_id,
    eos_token_id=tokenizer.eos_token_id,
    n_layer=12,
    n_head=12,
    n_embd=384
)

new_model=GPT2LMHeadModel(config)
```

```python
data_collator=DataCollatorForLanguageModeling(tokenizer=tokenizer,mlm=False)
```

```python
training_args=TrainingArguments(
    output_dir="./gpt2-Pretrained",
    num_train_epochs=400,
    per_device_train_batch_size=32,
    eval_steps=50,
    save_total_limit=2
)
```

```python
trainer=Trainer(
    model=new_model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=tokenized_data["train"]
)
trainer.train()
```

```python
new_model.save_pretrained("./gpt2-Pretrained_model")
```

```python
prompt="Who won 2025 IPL?"
tokenizer.save_pretrained("./gpt2-Pretrained")
from transformers import pipeline

text_generator = pipeline(
    "text-generation",
    model="./gpt2-Pretrained_model",
    tokenizer="./gpt2-Pretrained"
)
```

```python
output = text_generator(prompt)
print(output)
```
